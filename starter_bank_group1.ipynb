```
import pandas as pd

campaign = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv')
campaign.head()
```

```
yes_nos1 = campaign['default'].replace(['yes','no','unknown'], [1,0,0])
yes_nos2 = campaign['housing'].replace(['yes','no','unknown'], [1,0,0])
yes_nos3 = campaign['loan'].replace(['yes','no','unknown'], [1,0,0])



marital = campaign['marital'].replace(["divorced","married","single","unknown"], ["divorced","married","single","single"])

campaign['default'] = yes_nos1
campaign['housing'] = yes_nos2
campaign['loan'] = yes_nos3
campaign['marital'] = marital


campaign.head()
```
